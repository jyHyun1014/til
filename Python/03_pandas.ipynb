데이터프레임 기본 집계

```python
# 열 고유값 확인 # 결과는 배열 형태
print(df['MaritalStatus'].unique())

# 열 고유값 개수 확인 # 결과는 series 형태
print(df['MaritalStatus'].value_counts())
```

```python
# 열 합계 조회
print(df['MonthlyIncome'].sum())

# 열 최댓값 조회
print(df['MonthlyIncome'].max())

# 각 열 평균값 확인
print(df[['Age', 'MonthlyIncome']].mean())

# 각 열 중앙값 확인
print(df[['Age', 'MonthlyIncome']].median())
```

데이터프레임 정렬

```python
df.sort_index(ascending=False) # 인덱스 기준 정렬
df.sort_values(by='MonthlyIncome', ascending=False)  # 열 기준 정렬
df.sort_values(by=['JobSatisfaction', 'MonthlyIncome'], ascending=[True, False])
```

데이터프레임 조회

```python
display(df['Attrition'])    # 시리즈(1차원)로 조회

display(df[['Attrition']])  # 데이터프레임(2차원)으로 조회
```

```python
# .loc[행조건, 열이름]
# df.loc[:, ['Age', 'Gender']]     # 모든 행의 선택 열 반환
display(df.loc[df['Age'] > 30, 'Age'])    # 시리즈로 반환
display(df.loc[df['Age'] > 30, ['Age']])  # 데이터프레임으로 선택 열 반환
display(df.loc[df['Age'] > 30])           # 데이터프레임으로 모든 열 반환
```

```python
# 여러 조건
df.loc[(df['DistanceFromHome'] > 10) & (df['JobSatisfaction'] == 4)]  # and
df.loc[(df['DistanceFromHome'] > 10) | (df['JobSatisfaction'] == 4)]  # or
```

```python
# ==과 같음
df.loc[df['JobSatisfaction'].isin([1,4])]
df.loc[(df['JobSatisfaction'] == 1) | (df['JobSatisfaction'] == 4)]
```

```python
# [25, 30] 구간 사이의 값
df.loc[df['Age'].between(25, 30)]
df.loc[(df['Age'] >= 25) & (df['Age'] <= 30)]
df.loc[df['Age'].between(25, 30, inclusive='both')] # both(기본값), left, right, neither
```

데이터프레임 집계
- `df.groupby(’집계기준변수’, as_index=True,)[’집계대상변수’].집계함수()`

```python
# MaritalStatus 별 Age 평균 --> 시리즈
df.groupby('MaritalStatus', as_index=True)['Age'].mean()
```

```python
# MaritalStatus 별 Age 평균 --> 데이터프레임
df.groupby('MaritalStatus', as_index=True)[['Age']].mean()
```

```python
# as_index=False : 행 번호를 기반으로 한 정수 값이 인덱스로 설정
df.groupby('MaritalStatus', as_index=False)[['Age']].mean()
```

```python
# MaritalStatus 별 Age, MonthlyIncome 평균
df.groupby('MaritalStatus', as_index=False)[['Age','MonthlyIncome']].mean()
```

```python
df.groupby('MaritalStatus', as_index=False).sum()  # 전체 열
```

```python
df.groupby(['MaritalStatus', 'Gender'], as_index=False)[['Age','MonthlyIncome']].mean()
```

- `df.groupby( ).agg(['함수1','함수2', ...])  여러 함수로 한꺼번에 집계`

```python
# 집계를 한 번에 수행 가능
df.groupby('MaritalStatus', as_index=False)[['MonthlyIncome']].agg(['min','max','mean'])
```

```python
# 여러 열에 대한 각 열마다 다른 집계를 한 번에 수행 가능 # 딕셔너리 형태로 지정
# Age는 mean, MonthlyIncome는 max
df.groupby('MaritalStatus', as_index=False).agg({'Age':'mean','MonthlyIncome':'max'})
```
